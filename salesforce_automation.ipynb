```python
import csv
import requests
import time

base_url = "https://i8e43d5anj.execute-api.ap-south-1.amazonaws.com/latest"
endpoint = "/v1/candidate"

headers = {
    "Content-Type": "application/json",
    "service_ref": "b8ilcgos"
}

def send_request(endpoint, data):
    url = base_url + endpoint
    response = requests.post(url, headers=headers, json=data)
    return response

# Import CSV and send requests
csv_file_path = 'test_1.csv'
with open(csv_file_path, mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Create candidate
        response_create = send_request(endpoint, {"candidate": row})
        print(f"Status Code for Create Candidate: {response_create.status_code}")
        print(response_create.json())
        
        # Wait for 5 seconds
        #time.sleep(5)

        # Update source
        guid = response_create.json().get('data', {}).get('candidate', {}).get('guid')
        url_source = f'{base_url}/v1/candidate/{guid}'
        data_source = {
            "candidate": {
                "source": "tiq-crawler LinkedIn"
            }
        }
        response_source = send_request(url_source, data_source)
        print(f"Status Code for Update Source: {response_source.status_code}")
        print(response_source.json())
        
        # Wait for 5 seconds
        time.sleep(5)

        # Send email invitation
        url_invite = f'{base_url}/v1/candidate/{guid}/invite/email'
        data_invite = {
            "role": "Java Developer"
        }
        response_invite = send_request(url_invite, data_invite)
        print(f"Status Code for Email Invitation: {response_invite.status_code}")
        print(response_invite.json())

        # Wait for 5 seconds before processing the next entry
        time.sleep(5)
```

```python
import pandas as pd

file_path = 'out_salesforce_architect_dataminer.csv'
df = pd.read_csv(file_path)

column1 = 'First Name'
column2 = 'Last Name'

# Remove "'s" from 'Last Name' column
df[column2] = df[column2].str.replace("'s", "", regex=False)

# Create a new column by merging the specified columns
df['full_name'] = df[column1] + ' ' + df[column2]
df = df.drop([column1, column2], axis=1)

column_order = ['full_name'] + [col for col in df if col != 'full_name']
df = df[column_order]

# Update 'email_id' column
df['email_id'] = df['Email'].str.replace('@gmail.com', '@yopmail.com', regex=False)

# Drop unnecessary columns
df = df.drop(['Company', 'Title', 'Summary', 'Email'], axis=1)

df['job_role'] = "Salesforce Architect"
df['country'] = "India"
df['date'] = "2023-11-22"
df['source'] = "tiq-crawler LinkedIn"

df = df[['full_name', 'email_id', 'job_role', 'country', 'Profile URL', 'date', 'source']]
df = df.rename(columns={'Profile URL': 'Linkedin_url'})

df.to_csv('salesforce_architect_india.csv', index=False)
```
