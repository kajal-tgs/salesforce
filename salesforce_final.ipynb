```python
import pandas as pd

file_path = 'python_data_engineer.csv'
df = pd.read_csv(file_path)

column1 = 'First Name'
column2 = 'Last Name'

# Remove "'s" from 'Last Name' column
df[column2] = df[column2].str.replace("'s", "", regex=False)

# Create a new column by merging the specified columns
df['full_name'] = df[column1] + ' ' + df[column2]
df = df.dropna(subset=['Email'])
# Update 'email_id' column
df['email_id'] = df['Email'].str.replace('@[Gg]mail.com', '@yopmail.com', regex=True)

column_order = ['full_name'] + [col for col in df if col != 'full_name']
df = df[column_order]

# Drop unnecessary columns
df = df.drop(['Company', 'Title', 'Summary'], axis=1)

df['job_role'] = "Python Data Engineer"
df['country'] = "India"
df['date'] = "2023-12-12"
df['source'] = "tiq-crawler LinkedIn"

df = df[['full_name', 'email_id', 'job_role', 'country', 'Profile URL', 'date', 'source']]
df = df.rename(columns={'Profile URL': 'Linkedin_url'})

df.to_csv('python_data_engineer_india.csv', index=False)
```

```python
import csv
import requests 
import time
base_url = "https://i8e43d5anj.execute-api.ap-south-1.amazonaws.com/latest"
endpoint = "/v1/candidate"
headers = {
    "Content-Type": "application/json",
    "service_ref": "b8ilcgos"
} 
def send_request(candidate_data):
    url = base_url + endpoint
    response = requests.post(url, headers=headers, json={"candidate": candidate_data})
    return response
csv_file_path = 'python_data_engineer_india.csv'
with open(csv_file_path, mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        response = send_request(row)
        print(f"Status Code: {response.status_code}")
        print(response.json())

        response_data = response.json()
        guid = response_data.get('data', {}).get('candidate', {}).get('guid')

        if guid:
            print("GUID:", guid)
            url = f'{base_url}/v1/candidate/{guid}'
            headers = {
                'Content-Type': 'application/json',
                'service_ref': 'b8ilcgos'
            }

            data = {
                "candidate": {
                "source": "tiq-crawler LinkedIn"
                }
            }

            response = requests.post(url, json=data, headers=headers)

            print(response.status_code)
            print(response.json())  # If the response contains JSON data

            url = f'{base_url}/v1/candidate/{guid}/invite/email'
            headers = {
                'Content-Type': 'application/json',
                'service_ref': 'b8ilcgos'
            }

            data = {
                "role": "Python Data Engineer"
            }

            response = requests.post(url, json=data, headers=headers)

            print(response.status_code)
            print(response.json())  # If the response contains JSON data
            time.sleep(5)
        else:
            print("GUID not found in the response.")
```
